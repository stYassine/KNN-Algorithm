```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
training_set =pd.read_csv('Classified Data', index_col=0)
```

```python
training_set.head()
```

```python
training_set.tail(10)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler =StandardScaler()
```

```python
scaler.fit(training_set.drop('TARGET CLASS', axis=1))
```

```python
scaled_features =scaler.transform(training_set.drop('TARGET CLASS', axis =1))
```

```python
df_features =pd.DataFrame(scaled_features, columns=training_set.columns[:-1])
```

```python
df_features
```

```python
X =df_features
```

```python
X.shape
```

```python
y =training_set['TARGET CLASS']
```

```python
y.shape
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.3, random_state =101)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn =KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
predictions =knn.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
cm =confusion_matrix(predictions, y_test)
```

```python
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(predictions, y_test))
```

```python
error_rate =[]
for i in range(1, 40):
    knn =KNeighborsClassifier(i)
    knn.fit(X_train, y_train)
    predict_i =knn.predict(X_test)
    error_rate.append(np.mean(predict_i != y_test))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, 40), error_rate, color='blue', lineStyle='dashed', marker='o', markerFaceColor='red', markerSize=10)
plt.title('Error Rate vs K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
knn =KNeighborsClassifier(n_neighbors=18)
knn.fit(X_train, y_train)
predictions =knn.predict(X_test)

cm =confusion_matrix(predictions, y_test)
sns.heatmap(cm, annot=True)

print(classification_report(predictions, y_test))
```
